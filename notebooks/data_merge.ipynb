```python
df = tsv_data1
```

```python
# Split the 'freq,na_item,unit,geo\TIME_PERIOD' column if needed
if 'freq,na_item,unit,geo\\TIME_PERIOD' in df.columns:
    df[['freq', 'na_item', 'unit', 'geo']] = df['freq,na_item,unit,geo\\TIME_PERIOD'].str.split(',', expand=True)
    df.drop(columns=['freq,na_item,unit,geo\\TIME_PERIOD'], inplace=True)

# Rename year columns by stripping potential leading/trailing spaces or correcting names
df.columns = [col.strip() for col in df.columns]

# Melt the dataframe to long format
df_long = pd.melt(df, id_vars=['freq', 'na_item', 'unit', 'geo'], var_name='Year', value_name='GDP')

# Clean GDP values: remove characters like 'p', 'b', ':'
df_long['GDP'] = df_long['GDP'].replace({':': None, ' p': '', ' b': '', 'p': '', 'b': ''}, regex=True)
df_long['GDP'] = pd.to_numeric(df_long['GDP'], errors='coerce')  # Convert to numeric, coercing errors to NaN

# Output the first few rows to check the transformation
print(df_long.head())
```

```python
gdp_data = df_long
```

```python
# Step 1: Reshape GDP data by pivoting on the 'unit' column to create separate columns for each unit type
gdp_pivoted = gdp_data.pivot_table(index=["geo", "Year"], columns="unit", values="GDP", aggfunc='first').reset_index()

# Show the reshaped GDP data
gdp_pivoted.head()
```

```python
df = tsv_data2
```

```python
# Check if there's a need to split the first column into multiple column
# Check if there's a need to split the first column into multiple columns
if 'freq,unit,icha11_hc,geo\\TIME_PERIOD' in df.columns:
    df[['freq', 'unit', 'icha11_hc', 'geo']] = df['freq,unit,icha11_hc,geo\\TIME_PERIOD'].str.split(',', expand=True)
    df.drop(columns=['freq,unit,icha11_hc,geo\\TIME_PERIOD'], inplace=True)

# Ensure all column names are clean
df.columns = [col.strip() for col in df.columns]

# Melt the dataframe to long format
df_long = pd.melt(df, id_vars=['freq', 'unit', 'icha11_hc', 'geo'], var_name='Year', value_name='Health_Expenditure')

# Clean GDP values: remove characters like 'p', 'b', ':'
df_long['Health_Expenditure'] = df_long['Health_Expenditure'].replace({':': None, ' p': '', ' b': '', 'p': '', 'b': ''}, regex=True)
df_long['Health_Expenditure'] = pd.to_numeric(df_long['Health_Expenditure'], errors='coerce')  # Convert to numeric, coercing errors to NaN


```

```python
health_data = df_long
```

```python
# Step 1: Reshape Health Expenditure data by pivoting on the 'unit' column to create separate columns for each unit type
health_expenditure_pivoted = health_data.pivot_table(index=["geo", "Year"], columns="unit", values="Health_Expenditure", aggfunc='first').reset_index()

# Show the reshaped Health Expenditure data
health_expenditure_pivoted.head()
```

```python
amr_data = csv_data
```

```python
amr_data
```

```python
# Step 2: Reshape AMR Surveillance Data by pivoting on the 'Indicator' column to create separate columns for each indicator type
amr_pivoted = amr_data.pivot_table(index=["HealthTopic", "Population", "Time", "RegionCode", "RegionName", "Unit"], 
                                   columns="Indicator", values="NumValue", aggfunc='first').reset_index()

# Show the reshaped AMR data
amr_pivoted.head()
```

```python
# Split the "Population" column into "Bacteria" and "Drug_Class"
amr_pivoted[['Bacteria', 'Drug_Class']] = amr_pivoted['Population'].str.split('|', expand=True)

# Display the modified AMR data with new columns
amr_pivoted[['HealthTopic', 'Bacteria', 'Drug_Class', 'Time', 'RegionCode', 'RegionName', 'Unit'] + amr_pivoted.columns[7:].tolist()].head()
```

```python
# Pivot the "Unit" field to create separate columns for each unit type, consolidating counts and percentages
amr_unit_pivoted = amr_pivoted.pivot_table(index=['HealthTopic', 'Bacteria', 'Drug_Class', 'Time', 'RegionCode', 'RegionName'], 
                                           columns='Unit', 
                                           values=[col for col in amr_pivoted.columns if col not in ['HealthTopic', 'Population', 'Time', 'RegionCode', 'RegionName', 'Unit', 'Bacteria', 'Drug_Class']],
                                           aggfunc='first').reset_index()

# Flatten the multi-level column structure after pivoting
amr_unit_pivoted.columns = ['_'.join(col).strip() if col[1] else col[0] for col in amr_unit_pivoted.columns.values]

# Show the modified AMR data with new unit-based columns
amr_unit_pivoted.head()
```

```python
amr_unit_pivoted.to_csv("amr_data.csv")
health_expenditure_pivoted.to_csv("health_expenditure.csv")
gdp_pivoted.to_csv("gdp_data.csv")
```

```python
# Extract rows where Drug_Class is 'Data quality'
data_quality_df = amr_unit_pivoted[amr_unit_pivoted['Drug_Class'] == 'Data quality']

# Select relevant columns for the data quality table
data_quality_table = data_quality_df[['Year', 'geo', 'RegionName', 'Completeness age_%', 'Completeness gender_%']]

# Remove rows where Drug_Class is 'Data quality' from the main dataset
final_data_cleaned = amr_unit_pivoted[amr_unit_pivoted['Drug_Class'] != 'Data quality']

# Display the first few rows of the new data quality table and the cleaned main data
data_quality_table.head()
```

```python
# Add 'Bacteria' column to the data quality table
data_quality_table['Bacteria'] = data_quality_df['Bacteria']
```
